## Code to generate the required data: 

```python
# Imports pandas
import pandas as pd
# Imports matplotlib
import matplotlib.pyplot as plt
# Imports numpy
import numpy as np

#read data as dataframe
data = pd.read_csv('allmn.csv')

# removes samples that contain NANs from a subset of the data 
def drop_nan(col_names, data):
    # data    : original dataframe
    # col_ids : column ids for the clean(no-nan) sub_dataset
    sub_data = data[col_names].dropna()
    return(sub_data)
    

sub = drop_nan(data.columns,data)

#pleaing guilty
plea_guilty=sub[sub["plea"]==1]
#pleaing not guilty
plea_nguilty = sub[sub["plea"]==2]
#creating the pie plot and saving
    
#making subsets of the plea_guilty data by race
#only looking at 6 races since there is only 1 race value of 7 in the data (unknown)
pg_1=plea_guilty[plea_guilty["race"]==1]
pg_2=plea_guilty[plea_guilty["race"]==2]
pg_3=plea_guilty[plea_guilty["race"]==3]
pg_4=plea_guilty[plea_guilty["race"]==4]
pg_5=plea_guilty[plea_guilty["race"]==5]
pg_6=plea_guilty[plea_guilty["race"]==6]
#making subsets of the plea_nguilty data by race
png_1=plea_nguilty[plea_nguilty["race"]==1]
png_2=plea_nguilty[plea_nguilty["race"]==2]
png_3=plea_nguilty[plea_nguilty["race"]==3]
png_4=plea_nguilty[plea_nguilty["race"]==4]
png_5=plea_nguilty[plea_nguilty["race"]==5]
png_6=plea_nguilty[plea_nguilty["race"]==6]

# Dataframes for pie chart
df_1 = pd.DataFrame({'plea guilty by race':[len(pg_1),len(pg_2), len(pg_3), len(pg_4), len(pg_5), len(pg_6)]}, index=["white", "black", "am ind", "hispanic", "asian", "other"])
df_2 = pd.DataFrame({'plea not guilty by race':[len(png_1),len(png_2), len(png_3), len(png_4), len(png_5), len(png_6)]}, index=["white", "black", "am ind", "hispanic", "asian", "other"])
```

## Code to generate the visualization: 

```python
#creating the pie plot for pleaing guilty
plot = df_1.plot.pie(y='plea guilty by race', figsize=(7, 7))
```

```python
#creating the pie plot for pleaing not guilty
plot = df_2.plot.pie(y='plea not guilty by race', figsize=(7, 7))
```

## Code to add formating : 

If you would like the change/add to the formattingof the pie chart, here are some options:

 * Changing the formatting:
     * You make one or more slices of the pie-chart pop-out using the explode option. The number 0 is making them all stay in and changing the number to 1 pushes the slice further out.
     
     * You can also display the percentage along with each slice by using the following lines of code. 
        
     * Matplotlib gives you the option to get creative and make your pie-chart look as vibrant as possible. 

Pop out: 

```python
explode = (0, 1, 1, 0,0,1)
plot = df_2.plot.pie(y='plea not guilty by race', figsize=(7, 7),explode=explode )
```

Display Percentage:

```python
plot = df_2.plot.pie(y='plea not guilty by race', figsize=(7, 7), autopct='%1.1f%%', startangle=45)
```

Changing colors:

```python
colors = ( "orange", "cyan", "pink", 
          "grey", "yellow", "blue") 

plot = df_1.plot.pie(y='plea guilty by race', figsize=(7, 7), colors = colors)
```
